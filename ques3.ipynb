```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=461131e5-5af0-4e50-bee2-223fad1e00ca&pf_rd_r=MR7V08K1X0DNQ7XZY82X&pf_rd_s=center-1&pf_rd_t=60601&pf_rd_i=india.toprated&ref_=fea_india_ss_toprated_india_tr_india250_hd")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
titles = []
for i in soup.find_all('td', class_="titleColumn"):
    titles.append(i.text)
    
titles  
```

```python
years=[]
for i in soup.find_all('span', class_="secondaryInfo"):
    years.append(i.text)
    
years  
```

```python
ratings = []
for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    ratings.append(i.text)
    
ratings 
```

```python
df=pd.DataFrame({'Rank':titles,'Rating': ratings,'Years':years})
df
```
